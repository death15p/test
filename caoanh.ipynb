```python
import pandas as pd
import requests
import os
import time

# --- ĐIỀN THÔNG TIN TỪ GOOGLE VÀO ĐÂY ---
API_KEY = "AIzaSyCkrtndbWZUAkt9n8aORVzUecv10Kg2o-s"
CX = "b521f59feda854bf4"
# ---------------------------------------

# Đọc file CSV
df = pd.read_csv("list_50_loai_tra.csv") 
danh_sach_tra = df["Tên sản phẩm"].dropna().tolist()

# Tạo thư mục lưu ảnh
os.makedirs("Anh_Tra", exist_ok=True)
print("Bắt đầu khởi động Google API... 🚀")

for ten_tra in danh_sach_tra:
    if type(ten_tra) != str or ten_tra.strip() == "" or "TÍCH ĐI" in ten_tra:
        continue
        
    print(f"Đang tìm ảnh cho: {ten_tra}...")
    
    # Ép Google hiểu đây là tìm sản phẩm thực tế để lấy hình đẹp
    query = f"{ten_tra} sản phẩm"
    
    # URL gọi API của Google
    url = f"https://www.googleapis.com/customsearch/v1"
    params = {
        'q': query,
        'cx': CX,
        'key': API_KEY,
        'searchType': 'image', # Chỉ tìm kiếm hình ảnh
        'num': 1               # Chỉ lấy 1 kết quả đầu tiên (tốt nhất)
    }
    
    try:
        response = requests.get(url, params=params)
        data = response.json()
        
        # Thêm đoạn này để bắt lỗi trực tiếp từ Google
        if 'error' in data:
            print(f"-> 🚨 Google từ chối vì lỗi: {data['error']['message']}")
        # Nếu Google trả về kết quả bình thường
        elif 'items' in data:
            image_url = data['items'][0]['link']
            
            # Tải ảnh về
            img_data = requests.get(image_url, timeout=5).content
            safe_name = "".join([c for c in ten_tra if c.isalnum() or c==' ']).rstrip()
            
            with open(f"Anh_Tra/{safe_name}.jpg", 'wb') as handler:
                handler.write(img_data)
            print(f"-> Đã tải thành công!")
        else:
            print(f"-> Google phản hồi tốt nhưng không có tấm ảnh nào khớp.")
            
    except Exception as e:
        print(f"-> Gặp lỗi mạng khi tải: {e}")
```
